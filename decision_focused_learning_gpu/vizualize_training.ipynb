```python

import numpy as np 
import pandas as pd
import argparse
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

input_dir = 'results/'
d = {'model' : [], 'epoch' : [] , 'loss' : [] , 'exp_train' : [] , 'exp_test' : [], 'dni_train' : [] , 'dni_test' : [], 'avg_dp' : [] , 'lr' : []}
l = []
with open(input_dir + 'df_training.txt', 'r') as file : 
    
    model = -1
    for line in file : 
        line = line.replace('\n', '').split(' | ')
        if len(line) > 2 :
            if line[0] == '0' :
                model += 1 
            l.append([model] + line) 
df = pd.DataFrame(l).astype(float)

fig, ax = plt.subplots(3,3,figsize = (20,15))
for i,ax in enumerate(ax.reshape(-1)) :
    sns.lineplot(data = df, x=1, y=i+2, hue=0, ax=ax)
fig.show()
```
